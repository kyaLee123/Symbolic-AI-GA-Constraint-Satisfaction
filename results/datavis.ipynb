<h>This is just to visualize the data to pick a threshold eval if needs be</h>

```python
# Plot the output data to determine a reasonable eval threshold
import matplotlib.pyplot as plt
import csv

# Open the CSV file
with open("./LargeInput1_output.csv", "r") as csvfile:
    csvreader = csv.reader(csvfile)
    next(csvreader)
    
    # Initialize lists to store data
    iterationNumbers = []
    bestEvals = []
    ninetiethEvals = []
    medianEvals = []
    worstEvals = []
    
    counter = 0
    # Read each row from the CSV file
    for row in csvreader:
        iterationNumbers.append(counter)
        bestEvals.append(int(row[0]))
        ninetiethEvals.append(int(row[1]))
        medianEvals.append(int(row[2]))
        worstEvals.append(int(row[3]))
        counter += 1
        if(counter >= 5000):
            break

# Plot the data
a = plt.axes()
a.set_xlabel("Iteration")
a.set_ylabel("Eval")
a.plot(iterationNumbers, bestEvals, label="Best")
a.plot(iterationNumbers, ninetiethEvals, label="90th Percentile")
a.plot(iterationNumbers, medianEvals, label="Median")
a.plot(iterationNumbers, worstEvals, label="Worst", linewidth=0.5)
a.legend()

# Show the plot
plt.show()
```
